```python
from happytransformer import HappyTextToText
```

```python
happy_tt = HappyTextToText("T5", "t5-base")
```

```python
from datasets import load_dataset
```

```python
import pandas as pd
train_dataset=pd.read_csv("train.csv")
eval_dataset=pd.read_csv("eval.csv")
```

```python
 before_result = happy_tt.eval("eval.csv")
```

```python
print("Before loss:", before_result.loss)
```

```python
from happytransformer import TTTrainArgs

args = TTTrainArgs(batch_size=32, num_train_epochs=15)
happy_tt.train("train.csv", args=args)
```

```python
before_loss = happy_tt.eval("eval.csv")

print("After loss: ", before_loss.loss)
```

```python
from happytransformer import TTSettings

beam_settings =  TTSettings(num_beams=20, min_length=1, max_length=50)
```

```python
example_1 = "grammar: Isheen boru ganama dhufe ."
result_1 = happy_tt.generate_text(example_1, args=beam_settings)
print(result_1.text)
```

```python
import csv

with open('eval.csv', 'r') as csvfile:
   reader = csv.reader(csvfile)
   next(reader)
   for row in reader:
       first_value = row[0]
       result_1 = happy_tt.generate_text(first_value, args=beam_settings)
       print(first_value, "\t", result_1.text)
```
